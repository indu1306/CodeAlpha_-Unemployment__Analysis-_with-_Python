```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Unemployment in India.csv")

print(df.head())
print(df.info())
```

```python
# Check missing values
print(df.isnull().sum())

# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Convert date column to datetime
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Check dataset again
print(df.info())
```

```python
print(df.describe())
```

```python
avg_rate = df['Estimated Unemployment Rate (%)'].mean()
print("Average Unemployment Rate:", avg_rate)
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(
    x='Date',
    y='Estimated Unemployment Rate (%)',
    data=df
)

plt.title("Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.show()
```

```python
state_avg = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,8))
state_avg.plot(kind='bar')

plt.title("Average Unemployment Rate by State")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
covid_period = df[df['Date'] >= '2020-03-01']
pre_covid = df[df['Date'] < '2020-03-01']

print("Pre-Covid Average:",
      pre_covid['Estimated Unemployment Rate (%)'].mean())

print("Covid Period Average:",
      covid_period['Estimated Unemployment Rate (%)'].mean())
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(
    x='Date',
    y='Estimated Unemployment Rate (%)',
    data=df
)

plt.axvline(
    pd.Timestamp('2020-03-01'),
    color='red',
    linestyle='--',
    label='Covid Start'
)

plt.legend()
plt.title("Impact of Covid-19 on Unemployment")
plt.show()
```

```python
df['Month'] = df['Date'].dt.month_name()

monthly_avg = df.groupby('Month')['Estimated Unemployment Rate (%)'].mean()

plt.figure(figsize=(12,6))
monthly_avg.plot(kind='bar')

plt.title("Monthly Average Unemployment Rate")
plt.ylabel("Rate (%)")
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=np.number)

plt.figure(figsize=(8,6))
sns.heatmap(
    numeric_cols.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```
